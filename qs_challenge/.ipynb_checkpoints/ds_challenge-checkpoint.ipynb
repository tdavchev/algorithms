```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.neural_network import MLPClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn import linear_model
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import GaussianNB
from sklearn import svm
```

```python
df = pd.read_csv('data-science-challenge/data.csv')
```

```python
def areThereNoZeros(df):
    # are there any 0's in the document:
    print(df.loc[df['Feature 1'] == 0.0])
    print(df.loc[df['Feature 2'] == 0.0])
    print(df.loc[df['Feature 3'] == 0.0])
    print(df.loc[df['Feature 4'] == 0.0])
    print(df.loc[df['Feature 5'] == 0.0])
    
areThereNoZeros(df) # --> yes
# --> fill NaN with zeros then:
df = df.fillna(0)
```

```python
unique_user_id = set(df['User ID'])
```

```python
# make sure there is nothing weird going on...
mins_1 = [min([i for i in df[df['User ID'] == usr]['Feature 1'] if i != None]) for usr in unique_user_id]
maxes_1 = [max([i for i in df[df['User ID'] == usr]['Feature 1'] if i != None]) for usr in unique_user_id]


mins_2 = [min([i for i in df[df['User ID'] == usr]['Feature 2'] if i != None]) for usr in unique_user_id]
maxes_2 = [max([i for i in df[df['User ID'] == usr]['Feature 2'] if i != None]) for usr in unique_user_id]


mins_3 = [min([i for i in df[df['User ID'] == usr]['Feature 3'] if i != None]) for usr in unique_user_id]
maxes_3 = [max([i for i in df[df['User ID'] == usr]['Feature 3'] if i != None]) for usr in unique_user_id]


mins_4 = [min([i for i in df[df['User ID'] == usr]['Feature 4'] if i != None]) for usr in unique_user_id]
maxes_4 = [max([i for i in df[df['User ID'] == usr]['Feature 4'] if i != None]) for usr in unique_user_id]


mins_5 = [min([i for i in df[df['User ID'] == usr]['Feature 5'] if i != None]) for usr in unique_user_id]
maxes_5 = [max([i for i in df[df['User ID'] == usr]['Feature 5'] if i != None]) for usr in unique_user_id]
```

```python
stats = np.array([[mins_1, maxes_1], [mins_2, maxes_2], [mins_3, maxes_3], [mins_4, maxes_4], [mins_5, maxes_5]])
```

```python
titles = ["Feature 1", "Feature 2", "Feature 3", "Feature 4","Feature 5"]

print(np.array(stats).shape)
for i in range(5):
    fig = plt.figure()
    ax = fig.add_subplot(111)    # The big subplot
    ax.plot(stats[i][0], label="min")
    ax.plot(stats[i][1], label="max")
    ax.set_xlabel('User ID')
    ax.set_ylabel('Value')
    ax.set_title(titles[i])
    ax.legend()
plt.show()
plt.close()
```

```python
def loadData(path='data-science-challenge/data.csv'):   
    f = open(path)
    dataFrame = pd.read_csv(f)
    
    return dataFrame

def getTrain(dataFrame):
    # get all unique users -> 1000
    unique_user_id = set(df['User ID'])
    # get first 800 for train as I want to measure
    # generalisation on new users..
    count = int(len(unique_user_id) * 0.8)
    curr_count = 0
    train = 0
    test = 0
    for usr in unique_user_id:
        for pt in df[df['User ID'] == usr]:
            if curr_count <= count:
                train += 1
            else:
                test += 1

        if curr_count <= count:
            curr_count += 1
            
    return train

def fillWithMean(df):
    #let's fill nans with means per user
    # I am sure there is a better way to do this...
    dct = []
    for usr in unique_user_id:   
        mn_1 = df[df['User ID'] == usr]['Feature 1'].mean()
        feat_1 = df[df['User ID'] == usr]['Feature 1'].fillna(mn_1).tolist()
        
        mn_2 = df[df['User ID'] == usr]['Feature 1'].mean()
        feat_2 = df[df['User ID'] == usr]['Feature 1'].fillna(mn_2).tolist()
        
        mn_3 = df[df['User ID'] == usr]['Feature 1'].mean()
        feat_3 = df[df['User ID'] == usr]['Feature 1'].fillna(mn_3).tolist()
        
        mn_4 = df[df['User ID'] == usr]['Feature 1'].mean()
        feat_4 = df[df['User ID'] == usr]['Feature 1'].fillna(mn_4).tolist()
        
        mn_5 = df[df['User ID'] == usr]['Feature 1'].mean()
        feat_5 = df[df['User ID'] == usr]['Feature 1'].fillna(mn_5).tolist()
        
        feat_6 = df[df['User ID'] == usr]['Feature 6'].tolist()
        feat_7 = df[df['User ID'] == usr]['Feature 7'].tolist()
        for i in range(len(feat_1)):
            dct.append({
                'User ID': usr,
                'Feature 1': feat_1[i],
                'Feature 2': feat_2[i],
                'Feature 3': feat_3[i],
                'Feature 4': feat_4[i],
                'Feature 5': feat_5[i],
                'Feature 6': feat_6[i],
                'Feature 7': feat_7[i]
            })

    df = pd.DataFrame(dct)
    return df

def preProcessDataset(dataFrame, withFeat6=False, withUID=False):
    dataFrame = fillWithMean(dataFrame)

    if withFeat6:
        feat_F6 = dataFrame['Feature 6'].tolist()
        labelEncoderF6 = LabelEncoder()
        labelEncoderF6.fit(feat_F6)
        labelsF6 = labelEncoderF6.transform(feat_F6)
        dataFrame['Feature 6'] = pd.Series(labelsF6)
    else:
        dataFrame = dataFrame.drop(['Feature 6'], 1)
        
    if withUID:
        feat_uid = dataFrame['User ID'].tolist()
        labelEncoderUID = LabelEncoder()
        labelEncoderUID.fit(feat_uid)
        labelsUID = labelEncoderUID.transform(feat_uid)
        dataFrame['User ID'] = pd.Series(labelsUID)
    else:
        dataFrame = dataFrame.drop(['User ID'], 1)

    headers = list(dataFrame)

    # Store the labels separately
    YIndex = headers.index('Feature 7')
    datasetMatrix = dataFrame.as_matrix()
    
    return datasetMatrix, YIndex

def splitData(train, datasetMatrix, YIndex):
    datasetTrain = datasetMatrix[:train]
    datasetTrainWithoutLabels = np.delete(datasetTrain, YIndex, 1)

    labels = datasetTrain[:, YIndex]

    datasetTest = datasetMatrix[train:]
    datasetTestWithoutLabels = np.delete(datasetTest, YIndex, 1)

    trueLabels = datasetTest[:, YIndex]
    
    return datasetTrainWithoutLabels, labels, datasetTestWithoutLabels, trueLabels

def scaleData(datasetTrainWithoutLabels, datasetTestWithoutLabels):
    # scale only over the training data statistics*
    scaler = StandardScaler()
    scaler.fit(datasetTrainWithoutLabels)
    datasetTrainWithoutLabels = scaler.transform(datasetTrainWithoutLabels)
    datasetTestWithoutLabels = scaler.transform(datasetTestWithoutLabels) 
    
    return datasetTrainWithoutLabels, datasetTestWithoutLabels

def predict(datasetTrainWithoutLabels, labels, datasetTestWithoutLabels, trueLabels):
    knn = KNeighborsClassifier()
    knn.fit(datasetTrainWithoutLabels, labels)

    predLabels = knn.predict(datasetTestWithoutLabels)
    print("KNN: ", accuracy_score(trueLabels, predLabels))

    gnb = GaussianNB()

    predLabels = gnb.fit(datasetTrainWithoutLabels, labels).predict(datasetTestWithoutLabels)
    print("NB Gaus: ", accuracy_score(trueLabels, predLabels))

    logistic = linear_model.LogisticRegression(C=1e5)
    logistic.fit(datasetTrainWithoutLabels, labels)

    predLabels = logistic.predict(datasetTestWithoutLabels)
    print("Logistic: ", accuracy_score(trueLabels, predLabels))

    # LBFGS for full batch..
    clf = MLPClassifier(solver='lbfgs', alpha=1e-5,
                        hidden_layer_sizes=(5, 2), random_state=1)
    clf.fit(datasetTrainWithoutLabels, labels)

    predLabels = clf.predict(datasetTestWithoutLabels)
    print("FF NN: ",accuracy_score(trueLabels, predLabels))
    
    
    decTree = DecisionTreeClassifier(max_depth=15)
    decTree.fit(datasetTrainWithoutLabels, labels)
    
    predLabels = decTree.predict(datasetTestWithoutLabels)
    print("Decision Tree: ", accuracy_score(trueLabels, predLabels))

    clf_rf = RandomForestClassifier(max_depth=2, random_state=0)
    clf_rf.fit(datasetTrainWithoutLabels, labels)

    predLabels = clf_rf.predict(datasetTestWithoutLabels)
    print("Random Forrest: ",accuracy_score(trueLabels, predLabels))
    
    adaB = AdaBoostClassifier()
    adaB.fit(datasetTrainWithoutLabels, labels)
    
    predLabels = adaB.predict(datasetTestWithoutLabels)
    print("AdaBoost: ", accuracy_score(trueLabels, predLabels))
    
def build_report():
    # Future: It would be good if I scaled only Feats 2-5 without 1 (and especially 6) since 1 had
    # large max and small min so scaling might affect...
    # Additionally, come up with a way to utilise the user ID groups ?
    # Removing colinearity between points will also improve performance
    # Plot results ...
    
    # load params
    dataFrame = loadData()
    trainSize = getTrain(dataFrame)
    
    # try without scaling the data and without the text features
    print("---------------------------------------------------------------------")
    print("Predicting data without using text-based features")
    datasetMatrix, YIndex = preProcessDataset(dataFrame)
    datasetTrainWithoutLabels, labels, datasetTestWithoutLabels, trueLabels = splitData(trainSize, datasetMatrix, YIndex)

    predict(datasetTrainWithoutLabels, labels, datasetTestWithoutLabels, trueLabels)

    # what if I scaled the data ...
    print("---------------------------------------------------------------------")
    print("Scaling data without using text-based features")
    datasetTrainWithoutLabels, datasetTestWithoutLabels = scaleData(datasetTrainWithoutLabels, datasetTestWithoutLabels)

    predict(datasetTrainWithoutLabels, labels, datasetTestWithoutLabels, trueLabels)

    # Add text based feature
    print("---------------------------------------------------------------------")
    print("Add Feature 6")
    datasetMatrix, YIndex = preProcessDataset(dataFrame, withFeat6=True)
    datasetTrainWithoutLabels, labels, datasetTestWithoutLabels, trueLabels = splitData(trainSize, datasetMatrix, YIndex)

    predict(datasetTrainWithoutLabels, labels, datasetTestWithoutLabels, trueLabels)

    # Scale Data
    print("---------------------------------------------------------------------")
    print("Scale Data")
    datasetTrainWithoutLabels, datasetTestWithoutLabels = scaleData(datasetTrainWithoutLabels, datasetTestWithoutLabels)

    predict(datasetTrainWithoutLabels, labels, datasetTestWithoutLabels, trueLabels)
```

```python
build_report()
```
